```r
actor <- c()
actor2 <- c()
for ( i in 1:nrow(data)) {
    if("20337505" %in% unlist(str_split(data$배우[i],","))) {
        print(i)
    }
    actor <- append(actor,unlist(str_split(data$배우[i],",")))
    actor2 <- append(actor2,unlist(str_split(juy$배우[i],",")))
}

```

```r
actor_count <- data.frame(table(actor))
```

```r
actor_count <- actor_count[-1,]
```

```r
row.names(actor_count) <- c()
```

```r
colnames(actor_count) <- c("배우","횟수")
```

```r
actor_count <- arrange(actor_count, desc(횟수))
```

```r
actor_count <- cbind(actor_count, 점수=rep(0,nrow(actor_count)))
```

```r
actor_count
```

```r
for(i in 1:nrow(data)) {
    for (j in 1:nrow(actor_count)) {
        if(actor_count[j,1] %in% unlist(str_split(data$배우[i],","))) {
            if (actor_count[j,1] %in% unlist(str_split(juy$배우[i],","))) {
                if (data$cluster[i] == 3) {
                    actor_count[j,3] <- actor_count[j,3] + 6
                } else if (data$cluster[i] == 1) {
                    actor_count[j,3] <- actor_count[j,3] + 4
                } else if (data$cluster[i] == 2) {
                    actor_count[j,3] <- actor_count[j,3] + 2
                }
            } else {
                if (data$cluster[i] == 3) {
                    actor_count[j,3] <- actor_count[j,3] + 3
                } else if (data$cluster[i] == 1) {
                    actor_count[j,3] <- actor_count[j,3] + 2
                } else if (data$cluster[i] == 2) {
                    actor_count[j,3] <- actor_count[j,3] + 1
                }                
            }

        } 
    }
}
```

```r
actor_count <- arrange(actor_count, desc(점수))
```
