```r
samp <- createDataPartition(res$cluster, p=0.7, list=F)
```

```r
# 학습용 데이터 셋
data.tr <- res[samp,c(5,14)]

# 검증용 데이터셋
data.test <- res[-samp,c(5,14)]

# 문제 및 정답 만들기
x <- subset(data.test, select = -cluster)
y <- data.test$cluster
```

```r
# Model 만들기
model.nnet <- nnet(cluster ~ ., data=data.tr, size = 10) 
```

```r
# 검증

pred <- predict(model.nnet, x, type = "class") # class : 분류된것(객관식), raw : 주관식
```

```r
ntest <- subset(test, select = c(스크린수, 상영횟수, 감독, 배우, cluster, 감독_배우))
```

```r
head(ntest)
```

```r
str(ntest)
```

```r
ntest$감독 <- as.factor(ntest$감독)
ntest$배우 <- as.factor(ntest$배우)
ntest$cluster <- as.factor(ntest$cluster)
ntest$감독_배우 <- as.factor(ntest$감독_배우)
```

```r
pred_res <- c()

for (i in 1:100) {
    samp <- createDataPartition(ntest$cluster, p=0.7, list=F)
    # 학습용 데이터 셋
    data.tr <- ntest[samp,-c(1,2)]

    # 검증용 데이터셋
    data.test <- ntest[-samp,-c(1,2)]

    # 문제 및 정답 만들기
    x <- subset(data.test, select = -cluster)
    y <- data.test$cluster

    # Model 만들기
    model.nnet <- nnet(cluster ~ ., data=data.tr, size = 8) 

    # 검증

    pred <- predict(model.nnet, x, type = "class") # class : 분류된것(객관식), raw : 주관식

    pred_res <- append(pred_res,mean(pred == y))
}
mean(pred_res)
```

```r
table(pred, y)
```

- 감독, 배우 : 0.81
