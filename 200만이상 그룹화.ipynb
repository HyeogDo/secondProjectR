```r
newData <- c()
for (i in 1:nrow(ndata)) {
    for(j in 1:length(unlist(str_split(ndata$주연[i],",")))) {
        newData <- rbind(newData, cbind(ndata[i,1:6], 주연=unlist(str_split(ndata$주연[i],","))[j],ndata[i,8]))
    }
}
```

```r
for ( i in 1:nrow(newData) ){
    for (j in 1:nrow(actor)) {
        if (actor$배우명[j] == newData$주연[i]) {
            actor$점수[j] <- actor$점수[j] + as.numeric(newData$그룹화[i])
        }
    }
}
```

```r
ractor <- c()
for (i in 1:nrow(newData)) {
    if (newData$주연[i] %in% actors$배우명) {
        ractor <- append(ractor, i)
    }
}
```

```r
top_30 <- newData[ractor,]
```

```r
rownames(top_30) <- c()
top_30$영화명 <- factor(top_30$영화명)
top_30$감독 <- factor(top_30$감독)
top_30$주연 <- factor(top_30$주연)
top_30$그룹화 <- factor(top_30$그룹화)
```

```r
samp <- c()
for (i in 1:length(as.character(unique(newData$영화명)))) {
    temp <- c()
    for (j in 1:nrow(newData)) {
        if (newData$영화명[j] == as.character(unique(newData$영화명))[i]) {
            temp <- append(temp, j)
        }
    }
    samp <- append(samp, sample(temp, round(length(temp)*0.7)))
}
```

```r
# 학습용 데이터 셋
data.tr <- newData[samp,c(5,6,7)]

# 검증용 데이터셋
data.test <- newData[-samp,c(5,6,7)]

# 문제 및 정답 만들기
x <- subset(data.test, select = -그룹화)
y <- data.test$그룹화

# Model 만들기
model.nnet <- nnet(그룹화 ~ ., data=data.tr, size = 1) 

pred <- predict(model.nnet, x, type = "class")

mean(pred == y)
```
